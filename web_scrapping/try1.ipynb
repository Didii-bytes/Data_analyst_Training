```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Najib_Razak"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

infobox = soup.find("table", class_="infobox")

if infobox:
    rows = infobox.find_all("tr")

    for i, row in enumerate(rows, start=1):

        header = row.find("th")
        value  = row.find("td")

        print(f"Row {i}:")

        # Case 1: Row has both header + value
        if header and value:
            print("KEY:", header.text.strip())
            print("VAL:", value.text.strip())

        # Case 2: Row has only a header (e.g., section titles)
        elif header:
            print("HEADER ONLY:", header.text.strip())

        # Case 3: Row has images or empty rows
        elif value:
            print("VALUE ONLY:", value.text.strip())

        else:
            print("[Empty or image row]")

        print("-" * 40)
else:
    print("Infobox not found.")
```
